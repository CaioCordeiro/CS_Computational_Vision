```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
fig_name_edge_1 = "./images/Image_(1)"
fig_type_edge_1 = ".jpg"
result_path = "./result/"
img_edge_1 = cv.imread(fig_name_edge_1+fig_type_edge_1)
plt.imshow(img_edge_1)
plt.show()
```

```python
canny_1 = cv.Canny(img_edge_1, 150, 200)
plt.imshow(canny_1)
plt.show()
```

```python
cv.imwrite(result_path + "Image_(1)" + fig_type_edge_1, canny_1)
```

```python
from skimage.io import imsave
from skimage.filters import (threshold_otsu, threshold_niblack,
                             threshold_sauvola)
from skimage.color import rgb2gray
```
