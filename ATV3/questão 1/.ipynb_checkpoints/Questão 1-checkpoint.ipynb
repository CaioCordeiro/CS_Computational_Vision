```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
def pepper_noise(img):
    height = len(img)
    width = len(img[0])
    new_img = []
    new_img = img
    for i in range(height):
        for j in range(width):
            if random.randint(1, 100) <= 15:
                new_img[i][j] = 0

    return new_img
```

```python
def salt_pepper_noise(img):
    height = len(img)
    width = len(img[0])
    new_img = []
    new_img = img
    for i in range(height):
        for j in range(width):
            if random.randint(1, 100) <= 10:
                if random.randint(1,2) == 1:
                    new_img[i][j] = 0
                else: 
                    new_img[i][j] = 255
    return new_img
```

```python
def gaussian_noise(img):
    gauss = np.random.normal(10,30,img.size)
    gauss = gauss.reshape(img.shape[0],img.shape[1],img.shape[2]).astype('uint8')
    noise = img + gauss
    return noise
```

```python
fig_name = "image_(1)"
fig_type = ".jpg"
result_path = "./result/"
img = Image.open(fig_name+fig_type)
plt.imshow(img, cmap="gray")
```

```python
arr = np.array(img)
pepper = pepper_noise(arr) 
img2 = Image.fromarray(pepper)
plt.imshow(img2, cmap="gray")
```

```python
arr2 = np.array(img)
salt_pepper = salt_pepper_noise(arr2) 
img3 = Image.fromarray(salt_pepper)
plt.imshow(img3, cmap="gray")
```

```python
arr3 = np.array(img)
gaussian = gaussian_noise(arr3) 
img4 = Image.fromarray(gaussian)
plt.imshow(img4, cmap="gray")
```

```python
img2.convert("L").save(result_path+fig_name+"_pepper"+fig_type)
img3.convert("L").save(result_path+fig_name+"_salt_pepper"+fig_type)
img4.convert("L").save(result_path+fig_name+"_gaussian"+fig_type)
```
