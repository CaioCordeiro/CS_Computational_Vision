```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_image(fig_name, fig_type, result_path, image):
    plt.imsave(result_path+fig_name+fig_type, image, cmap='gray')
    plt.imshow(image, cmap='gray')
```

```python
def resize(img, new_height, new_width):
    height = len(img)
    width = len(img[0])
    return [[ im[int(height * r / new_height)][int(width * c / new_width)]  
                 for c in range(new_width)] for r in range(new_height)]
```

```python
fig_name = "Zoom_in_(1)"
fig_type = ".jpg"
result_path = "./result/"
im = plt.imread(fig_name+fig_type)
new_img = resize(im, 480, 360)
plot_image(fig_name, fig_type, result_path, new_img)
```
