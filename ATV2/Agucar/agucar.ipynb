```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
def convolve2D(image, kernel):
    m, n = kernel.shape
    if (m == n):
        h, w = image.shape
        h = h - m + 1
        w = w - m + 1
        new_image = np.zeros((h,w))
        for i in range(h):
            for j in range(w):
                new_image[i][j] = np.sum(image[i:i+m, j:j+m]*kernel)
    return new_image
```

```python
# sharpen = np.array([[-1, -1 , -1],
#                     [-1, 9, -1],
#                     [-1, -1, -1]])
sharpen = np.array([[-1, -1 , 0],
                    [-1, 7, -1],
                    [0, -1, -1]])
# sharpen = np.array([[0, -4 , 0],
#                     [0, 5, 0],
#                     [0, 0, 0]])

image = cv2.imread('./Agucar_(1).jpg', cv2.IMREAD_GRAYSCALE)
image_sharp = convolve2D(image, sharpen)
plt.imshow(image, cmap='gray'), plt.axis("off")
plt.show()
plt.imshow(image_sharp, cmap='gray'), plt.axis("off")
plt.show()
to_save = Image.fromarray(image_sharp).convert("L")
to_save = to_save.save("./result/Agucar_(1).jpg")
```
